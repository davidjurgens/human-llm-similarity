```python
with jsonlines.open('data/wildchat_subset_en_100k_all_normalized_metrics.jsonl') as reader:
    data = list(reader)
df_dep = pd.DataFrame(data)
```

```python
df_dep
df_dep.columns.tolist()
df_dep = df_dep.loc[:, ~df_dep.columns.str.startswith(('agg_', 'metric_')) & (df_dep.columns != 'overall_score')]
df_dep
```

```python
combined_metrics_with_dv = pd.merge(df_dep, df_metrics, on=['conversation_hash', 'hashed_ip'], how='left') 
```

```python
combined_metrics_with_dv
```

```python
#unfold all dictionary columns
#which have type dict
for column in combined_metrics_with_dv.columns:
    if combined_metrics_with_dv[column].apply(type).eq(dict).all():
        print(f"Column '{column}' has dtype 'dict'")
#print(result_df['llm_punctuation'][0]) sanity chcek I thought I went insane
```

```python
df_combined = combined_metrics_with_dv.copy()

columns_to_expand = [
    'ai_turn_2_liwc_x', 'human_turn_1_liwc_x', 'ai_turn_2_punctuation_x', 'human_turn_1_punctuation_x',
    'ai_turn_2_pos_x', 'human_turn_1_pos_x'
]

for col in columns_to_expand:
    # Normalize the JSON objects in the specified column
    df_expanded = pd.json_normalize(combined_metrics_with_dv[col])
    
    # Create new column names by prefixing with the original column name and appending the JSON object keys
    df_expanded.columns = [f'{col}_{subcol}' for subcol in df_expanded.columns]
    
    # Drop the original column and concatenate the expanded DataFrame with the original DataFrame
    df_combined = pd.concat([df_combined.drop(columns=[col]), df_expanded], axis=1)
```
