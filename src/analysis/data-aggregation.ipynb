```python
def check_human_turn_1_stays_same(group):
    return len(group['human_turn_1'].unique()) == 1

group_sizes = df_metrics.groupby('conversation_hash').size()
unique_group_sizes = group_sizes.nunique()
print(len(set(group_sizes)))
total_groups = len(group_sizes)
unique_sizes = group_sizes.unique()


print(f"Unique group sizes: {unique_sizes}")
print(f"number of groups: {total_groups}")

grouped = df_metrics.groupby('conversation_hash')
consistency = grouped.apply(check_human_turn_1_stays_same)
inconsistent_groups = consistency[~consistency].index
print(f"Groups where 'human_turn_1' does not stay the same: {inconsistent_groups.tolist()}")

#TO DO: CHECK IF SAME - METRICS
```

```python
df_combined = result_df.copy()

# List of columns to expand
columns_to_expand = [
    'ai_turn_2_liwc', 'human_turn_1_liwc', 'ai_turn_2_punctuation', 'human_turn_1_punctuation',
    'ai_turn_2_pos', 'human_turn_1_pos', 'human_liwc', 'human_pos', 'human_punctuation',
    'llm_liwc', 'llm_pos', 'llm_punctuation'
]

# Loop through each column, expand it, and concatenate to df_combined
for col in columns_to_expand:
    df_expanded = pd.json_normalize(result_df[col])
    df_combined = pd.concat([df_combined.drop(columns=[col]), df_expanded], axis=1)
```
