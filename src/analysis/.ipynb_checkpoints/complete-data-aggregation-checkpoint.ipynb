```python
print(combined_metrics_with_dv['human_turn_1_topic'])
num_categories = 15
for i in range(num_categories):
    combined_metrics_with_dv[f'category_{i}'] = combined_metrics_with_dv['human_turn_1_topic'].apply(
        lambda x: x[i] if isinstance(x, list) and i < len(x) else None
    )
print(combined_metrics_with_dv['human_turn_1_topic'][0])
combined_metrics_with_dv.columns.tolist()
```

```python
combined_metrics_with_dv.columns.tolist()
```

```python
#unfold all dictionary columns
#which have type dict
columns_to_expand = []
for column in combined_metrics_with_dv.columns:
    if combined_metrics_with_dv[column].apply(type).eq(dict).any():
        print(f"Column '{column}' has dtype 'dict'")
        columns_to_expand.append(column)
#print(result_df['llm_punctuation'][0]) sanity chcek I thought I went insane
```

```python
print(columns_to_expand)
```

```python
df_combined = combined_metrics_with_dv.copy()

for col in columns_to_expand:

    df_expanded = pd.json_normalize(combined_metrics_with_dv[col])
    
    df_expanded.columns = [f'{col}_{subcol}' for subcol in df_expanded.columns]
    
    df_combined = pd.concat([df_combined.drop(columns=[col]), df_expanded], axis=1)
```
