```python
df = pd.read_json("data/english_only/prompting_results_clean/with_metrics/wildchat_subset_en_2k_prompting_Meta-Llama-3.1-70B-Instruct.jsonl", orient='records', lines=True)
df = df[df.llm_turn_3 != 'INVALID_DO_NOT_USE]']
```

```python
avg_df = df[["prompt"] + [col_name for col_name in df.columns 
                          if "metric_" in col_name]].groupby("prompt").mean()

print(avg_df.head())
print()
print(avg_df.rank(ascending=False).mean(axis=1).sort_values())
```

```python
import seaborn as sns
sns.set_theme(rc={'figure.figsize':(11.7,8.27)})

plot_df = pd.melt(avg_df.reset_index(), id_vars = ['prompt'], value_vars = avg_df.columns)
ax = sns.pointplot(data = plot_df, y = 'prompt', x = 'value', hue = 'variable')
sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))
```

```python
cot_prompts = [5,9,10,12,17,19,25,26,28,30,31,33,46]
weird_prompts = [28,33,35,37,38,40,41,42,43,45,47,50]
override_prompts = [11,22,29,48,49]
override_prompts.extend(weird_prompts)
human_prompts = [i for i in range(1,51) 
                 if i not in cot_prompts and i not in weird_prompts and i not in override_prompts]
```

```python
for title, keep_prompts in {'COT':cot_prompts, 'OVERRIDE': override_prompts, 'HUMAN':human_prompts}.items():
    rank_df = avg_df.drop(['metric_liwc','metric_formality','metric_politeness'], axis=1)
    for c in rank_df.columns:
        rank_df[c] = (-rank_df[c]).rank()

    sns.set_theme(rc={'figure.figsize':(5,4)})
    plot_df = pd.melt(rank_df.transpose().reset_index(), 
                      value_vars = [c for c in rank_df.index if c!='prompt'], value_name = 'rank')
    plot_df = plot_df[plot_df.prompt.isin(['Prompt_'+str(i) for i in keep_prompts])]

    import matplotlib.pyplot as plt
    color = plot_df.prompt == plot_df.groupby('prompt').quantile(q=0.75).sort_values('rank').index[0] 
    sns.boxplot(data = plot_df, y = 'prompt', x = 'rank', hue = color,
                order = plot_df.groupby('prompt').quantile(q=0.75).sort_values('rank').index, legend=False)
    plt.xlabel('Rank')
    plt.ylabel('')
    plt.title(title)
    plt.savefig('/home/akananth/Misc/prompt_ranks_'+title+'.pdf', bbox_inches='tight')
    plt.show()
```
